```python
import pandas
import numpy as np
import tensorflow as tf
```

```python
dataset = pandas.read_csv('https://covid-prediction-dataset.s3.ap-south-1.amazonaws.com/Cleaned-Data.csv')
```

```python
dataset
```

```python
dataset.columns
```

```python
y = dataset[['Severity_Severe']]
```

```python
X = dataset[['Fever', 'Tiredness', 'Dry-Cough', 'Difficulty-in-Breathing',
       'Sore-Throat', 'Pains', 'Nasal-Congestion',
       'Runny-Nose', 'Diarrhea', 'Age_0-9', 'Age_10-19',
       'Age_20-24', 'Age_25-59', 'Gender_Female',
       'Gender_Transgender', 'Contact_Dont-Know','Contact_Yes']]
```

```python
X = np.asarray(X).astype(np.float32)
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=62)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from keras.models import Sequential
model = Sequential()
from keras.layers import Dense
```

```python
#model.add(Dense(bias_initiunits = 27, input_dim = 17, activation = 'relu'))
model.add(Dense(kernel_initializer='Zeros', bias_initializer='Zeros', units = 72, input_dim = 17, activation = 'relu'))
model.add(Dense(units = 22,   activation = 'relu'))
model.add(Dense(units = 11,   activation = 'relu'))
model.add(Dense(units = 55,   activation = 'relu'))
#model.add(Dense(units = 10,   activation = 'relu'))
```

```python
model.add(Dense(units = 2, activation = 'softmax'))
from keras.optimizers import RMSprop
model.compile(optimizer= RMSprop(learning_rate=0.00003), loss = 'categorical_crossentropy' , metrics=['accuracy'])
from keras.optimizers import Adam
model.compile(optimizer= Adam(), loss = 'mean_squared_error' , metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=150)
model.get_weights()
print(model.history)
```

```python
model.evaluate(X_test , y_test)
```
